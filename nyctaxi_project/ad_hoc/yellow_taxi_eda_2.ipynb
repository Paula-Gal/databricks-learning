```python
from pyspark.sql.functions import date_format, count, sum
```

```python
spark.read.table("nyctaxi.`01_bronze`.yellow_trips_raw").groupBy(date_format(
    "tpep_pickup_datetime", "yyyy-MM-dd").alias("year_month")).agg(
    count("*").alias("total_records")).orderBy("year_month").display()
```
