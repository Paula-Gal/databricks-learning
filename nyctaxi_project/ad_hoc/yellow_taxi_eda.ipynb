```python
# Exploratory Data Analysis

# Which vendor makes the most revenue?
# What is the most popular pickup borough?
# Waht is the most common journey (borough to borough)?
# Create a time series chart showing the number of trips and and total revenue per day
```

```python
# 1. Which vendor makes the most revenue?
df_silver_enriched = spark.read.table("nyctaxi.02_silver.yellow_trips_enriched")
df_gold_summary = spark.read.table("nyctaxi.03_gold.daily_trips_summary")
```

```python
df_silver_enriched.display()
df_gold_summary.display()
```

```python
from pyspark.sql.functions import sum, max, round, count
```

```python
df_grouped=df_silver_enriched.groupBy("vendor").agg(sum("total_amount")).alias("total_amount")
```

```python
df_grouped = df_grouped.withColumnRenamed(
    "sum(total_amount)", "total_amount"
)
```

```python
df_grouped.display()
```

```python
df_grouped.select(max("total_amount")).alias("final_vendor").show()
```

```python
df_silver_enriched.display()
```

```python
df_silver_enriched.groupBy("vendor").agg(sum("total_amount")).alias("total_amount").withColumnRenamed(
    "sum(total_amount)", "total_amount"
).select(max("total_amount")).alias("final_vendor").show()
```

```python
df_silver_enriched.groupBy("vendor").agg(sum("total_amount").alias("total_amount")).orderBy("total_amount", ascending=False).limit(1).display()
```

```python
# 2. What is the most popular pickup borough?
df_silver_enriched.groupBy("pu_borough").agg(count("*").alias("no_trips")).orderBy("no_trips", ascending=False).display()
```

```python
# 3. The most common journey
df_silver_enriched.groupBy("pu_borough", "do_borough").agg(count("*").alias("no_trips")).orderBy("no_trips", ascending=False).display()
```
