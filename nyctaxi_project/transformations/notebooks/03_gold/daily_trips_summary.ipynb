```python
from pyspark.sql.functions import count, max, min, avg, sum, round
```

```python
df = spark.read.table("nyctaxi.02_silver.yellow_trips_enriched")
```

```python
df.display()
```

```python
df = df.groupBy(
    df.tpep_pickup_datetime.cast("date").alias("pickup_time")).agg(
        count('*').alias("total_trips"),
        round(avg("passenger_count"), 1).alias("avg_passenger_count"),
        round(avg("trip_distance"), 1).alias("avg_trip_distance"),
        round(avg("fare_amount"), 2).alias("avg_fare_amount"),
        max("fare_amount").alias("max_fare"),
        min("fare_amount").alias("min_fare"),
        round(sum("total_amount"), 2).alias("total_revenue"))
```

```python
df.display()
```

```python
df.write.mode("overwrite").saveAsTable("nyctaxi.03_gold.daily_trips_summary")
```
