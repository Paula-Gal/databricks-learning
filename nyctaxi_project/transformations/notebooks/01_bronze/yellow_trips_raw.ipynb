###This notebook will create a table with the same name in the bronze layer

```python
from pyspark.sql.functions import current_timestamp
```

```python
df = spark.read.format("parquet").load("/Volumes/nyctaxi/00_landing/data_sources/nyctaxi_yellow/*")
```

```python
df.display()
```

```python
df = df.withColumn("processed_timestamp", current_timestamp())
```

```python
df.write.mode("overwrite").saveAsTable("nyctaxi.01_bronze.yellow_trips_raw") # saved as a managed delta table
```
