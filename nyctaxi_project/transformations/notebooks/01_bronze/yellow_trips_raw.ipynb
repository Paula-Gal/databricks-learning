###This notebook will create a table with the same name in the bronze layer

```python
from pyspark.sql.functions import current_timestamp
from dateutil.relativedelta import relativedelta
```

```python
# Obtain the year-month for 2 months prior to the current month in yyyy-MM format
two_months_ago = date.today() - relativedelta(months=2)
formatted_date = two_months_ago.strftime("%Y-%m")

# Real all Parquet file for the specified month from the landing directory into a Data Frame
df = spark.read.format("parquet").load(f"/Volumes/nyctaxi/00_landing/data_sources/nyctaxi_yellow/{formatted_date}")
```

```python
df = df.withColumn("processed_timestamp", current_timestamp())
```

```python
# Write the Data Frame to a Unity catalog managed Delta table in the bronze schema, appending new data
df.write.mode("append").saveAsTable("nyctaxi.01_bronze.yellow_trips_raw") # saved as a managed delta table
```
