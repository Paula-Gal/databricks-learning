```python
from pyspark.sql.functions import lit, upper
```

```python
dt.update(
          condition = "name = 'Asia'",
          set = {
              "id": lit(100),
              "name": upper("name")
          }
        )
```

```python
df = spark.read.format("csv").option("header", True).schema("id int, name string").load("/Volumes/population_metrics/landing/datasets/countries_dataset/csv_data/country_regions/")

df.display()
```

```python
dt.alias("target").merge(
    df.alias("source"),
    "target.id = source.id"
).whenMatchedUpdate(
    set = {
        "target.name": "source.name"
    }
).whenNotMatchedInsert(
    values= { 
             "target.id": "source.id",
             "target.name": "source.name"
    }
).execute()
```
